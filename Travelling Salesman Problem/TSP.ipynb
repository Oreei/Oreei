##**Présentation du problème du voyageur de commerce**

Avant d’entamer la résolution de ce problème, nous allons le présenter.
Il s’agit d’un des problèmes d’optimisation combinatoire les plus célèbres qui existe depuis le XIXème siècle. L’énoncé de ce problème est le suivant :

*« Un voyageur de commerce doit visiter une et une seule fois un nombre fini de villes et revenir à son point d’origine. Trouvez l’ordre de visite des villes qui minimise la distance totale parcourue par le voyageur ».*
**William Rowan Hamilton**

Ce problème appartient à la classe des problèmes « NP-difficile » c’est-à-dire qu’on ne connait pas d’algorithme exacte polynomial.

Pour rappel, un algorithme polynomial est un algorithme qui va trouver la solution optimale à un problème et dont son temps d’exécution est une fonction polynomiale de degré n dépendante de la taille des données d’entrée. Ici n correspond à une constante entière.

Aux premiers abords, le problème pourrait sembler trivial. Cependant si le nombre de ville à visiter est trop important, alors cela peut engendrer une explosion combinatoire pouvant rendre la résolution impossible pour nos ordinateurs actuelles.

Pour comprendre cela, imaginons que n villes soient sélectionnées pour ce problème. Le nombre de chemins possibles est donc égale à $$\frac{{(n-1)!}}{2}$$ .

Cela s’explique par le fait que :

	- le point de départ est choisi de fonction arbitraire et n’influence en rien la longueur du chemin. Il est alors retranché au nombre de ville pouvant créer un chemin
	- un chemin peut être parcouru de deux manière différente, sa distance sera identique. On divise donc le nombre de chemin par 2

Si 5 villes sont sélectionnées, il existera 12 chemins possible. En revanche si 100 villes sont sélectionnées, il existera 4.666×10^155 chemins possibles.

A l’heure actuelle, il n’existe pas de méthode exacte et donc pas de constante n pour le problème du voyageur de commerce. Les problèmes « NP-difficile » sont caractérisés de la même manière mais pour autant on les retrouve dans de nombreux domaine d’application dont la chimie.

Une étude a, par exemple, démontré que l’utilisation des algorithme d’optimisation combinatoire guidé par l’intelligence artificiel pouvait générer des molécules pouvant satisfaire des propriétés moléculaires bien spécifique.
Le problème du voyageur de commerce étant un algorithme de combinaison combinatoire, nous pourrions alors envisager son utilisation dans la chimie moléculaire.

Le sujet ayant été présenté, nous allons maintenant le résoudre via différentes méthodes d’optimisation.
Pour rappel, le total trajet doit être le plus court possible et doit satisfaire les conditions suivantes :

	- passer par toutes les villes sélectionnées
	- ne jamais passer deux fois par le même point
	- démarrer par la ville la plus peuplée du jeu de donnée à savoir Shanghai



##**Identification de toutes les permutations**

Dans le contexte du Problème du Voyageur de Commerce, la première méthode qui nous vient à l'esprit est l'identification de toutes les permutations. Nous allons explorer cette approche directe, qui consiste à générer toutes les permutations possibles des villes à visiter, à calculer la distance totale pour chaque permutation, et enfin à sélectionner celle qui minimise cette distance. Notre objectif est de trouver une solution optimale pour un ensemble maximal de villes, en utilisant des données géographiques telles que les coordonnées de latitude et de longitude pour calculer les distances.

###**Formalisme mathématique**

Dans cette section, nous allons présenter le formalisme mathématique qui se cache derrière cette approche. Ceci fournira une base solide pour comprendre les détails techniques et mathématiques de la méthode.

Soit V un ensemble fini de villes.

*V = {v1, v2, …, vn}*

V représente l'ensemble de toutes les villes que nous souhaitons visiter.

Soit d une fonction qui donne la distance entre deux villes.

*d : V × V → R+*

La fonction d prend deux villes comme entrée et retourne un nombre positif (la distance entre elles). Par exemple, d(vi, vj) représente la distance entre les villes vi et vj. Dans le contexte de villes sur Terre, cela pourrait être la distance haversine tenant compte de la courbure de la Terre.

Soit P l'ensemble de toutes les permutations possibles des villes.

*P = {p1, p2, …, p(1/2)(n−1)!}*

Chaque représente une séquence unique des villes. Étant donné que nous avons fixé une ville comme point de départ et que chaque trajet a un inverse, le nombre total de permutations uniques est (1/2)(n−1)!

Pour chaque permutation, nous calculons alors la longueur totale du trajet.

Pour se faire, nous additionnons les distances entre chaque paire de villes consécutives dans la permutation pk :

***L(pk) = ∑(de i=1 à n−1) d(pk(i), pk(i+1))***

**La permutation qui aura la plus courte longueur totale de trajet sera notre solution optimale popt.**

***popt = arg⁡ min{pk ∈ P} (L(pk))***

où arg min est la fonction retournant la permutation popt pour laquelle L(pk) est minimale, c'est-à-dire le trajet le plus court.

Le formalisme basé sur la méthode des permutations capture l'essence exhaustive du Problème du Voyageur de Commerce. Il transforme la complexité du problème en équations précises, démontrant la rigueur de cette approche.

````python
import time
from itertools import permutations
from geopy.distance import great_circle
````

Importons maintenant les modules nécessaires pour résoudre le Problème du Voyageur de Commerce via la méthode d'identification des permutations. Le module time nous permettra de mesurer le temps d’exécution, tandis que itertools nous aidera à générer les permutations des villes à visiter. geopy.distance nous offre quant à elle la fonction great_circle pour calculer les distances géographiques réalistes entre les villes. Ces outils sont essentiels dans notre approche pour résoudre le TSP.

````python
import pandas as pd
import time

t1 = time.time()

# Lien direct pour télécharger le fichier depuis Google Drive
url = "https://drive.google.com/uc?export=download&id=1clgBjvEbQdrq43k1H9tm8jGy96cJC2E4"

# Charger le jeu de données
data = pd.read_csv(url, delimiter=';')
data = data.sort_values(by="Population", ascending=False)
# Sélectionner seulement les données des 10 plus grandes villes
data = data.iloc[:10, :]

# Extraire les coordonnées des villes
coordinates = data["Coordinates"].str.split(",", expand=True).astype(float)
````

Ici, nous enregistrons le temps de début d’exécution pour surveiller la performance de notre programme. Ensuite, nous chargeons les données à partir d’un fichier CSV, les triant par ordre décroissant de population pour mettre en évidence les villes les plus peuplées. Ensuite, nous sélectionnons les 10 plus grandes villes à partir de ces données triées.

La partie la plus importante de cette section consiste à extraire les coordonnées géographiques de ces 10 villes. Cela se fait en divisant les coordonnées en latitude et longitude, puis en les convertissant en valeurs flottantes pour les rendre exploitables.

En combinant ces étapes, nous créons un ensemble de données prêt à être utilisé pour résoudre le Problème du Voyageur de Commerce en utilisant les coordonnées géographiques réelles de ces villes.

````python
def calcul_longueur(a, b):
    return great_circle(a, b).km
````

On définit une fonction `calcul_longueur(a, b)`. Elle permet de calculer la distance réelle en kilomètres entre deux points géographiques `a` et `b`. Cette fonction utilise la formule de la distance `great_circle` du module `geopy.distance`, qui prend en compte la courbure de la Terre pour fournir des mesures précises.

Lorsque cette fonction est appelée avec deux ensembles de coordonnées géographiques, elle renvoie la distance en kilomètres entre ces deux points.

````python
# Ville de départ (Shanghai)
start_city = coordinates.iloc[0].values

# Générer toutes les permutations possibles des villes (sauf la ville de départ)
cities = coordinates.iloc[1:].values
permutations_list = permutations(cities)

# Initialiser la distance minimale à une valeur élevée
min_distance = float('inf')
best_path = None
````

Nous déterminons notre ville de départ, Shanghai dans ce cas. Ensuite, nous générons toutes les permutations possibles des autres villes (à l'exception de la ville de départ). Ces permutations représentent différents ordres de visite des villes.

Pour évaluer la meilleure solution, nous initialisons la distance minimale à une valeur élevée (l'infini) et la variable `best_path` à `None`. Ces variables seront utilisées pour suivre la distance la plus courte trouvée jusqu'à présent et le meilleur chemin, respectivement.

Cette étape préparatoire est cruciale pour résoudre efficacement le Problème du Voyageur de Commerce, car elle crée un ensemble de permutations à explorer pour trouver la solution optimale. Nous sommes maintenant prêts à passer à l'étape suivante, où nous évaluerons ces permutations pour trouver le meilleur chemin.

````python
# Calculer la distance pour chaque permutation et trouver le chemin le plus court
for path in permutations_list:
    path_distance = 0
    prev_city = start_city
    for i, city in enumerate(path):
        path_distance += calcul_longueur(prev_city, city)
        prev_city = city

    if path_distance < min_distance:
        min_distance = path_distance
        best_path = [start_city] + list(path)
````

Ensuite, nous explorons toutes les permutations possibles des villes à visiter, à l’exception de la ville de départ qui est Shanghai. Notre objectif est de trouver le chemin le plus court parmi toutes ces permutations.

À chaque itération de la boucle, nous calculons la distance totale parcourue pour la permutation actuelle. Pour ce faire, nous accumulons les distances entre chaque paire de villes consécutives en utilisant la fonction calcul_longueur.

Nous comparons ensuite cette distance calculée avec la distance minimale trouvée jusqu’à présent. Si la distance de la permutation actuelle est plus courte, cela signifie que nous avons trouvé un chemin plus court, et nous mettons à jour la distance minimale (min_distance) ainsi que le meilleur chemin (best_path) pour refléter cette découverte.

Ce processus se répète pour chaque permutation, nous permettant de trouver le chemin le plus court parmi toutes les combinaisons possibles. À la fin de cette boucle, best_path contiendra le meilleur ordre de visite des villes, et min_distance indiquera la distance totale parcourue pour ce chemin. C’est la solution optimale au Problème du Voyageur de Commerce en utilisant la méthode des permutations.

Pour être plus précis, nous avons tracé ci-dessous une courbe montrant le nombre de ville, noté n, en fonction du nombre de permutation calculé à partir de la formule (n-1)! / 2. On observe une croissance exponentielle du nombre de permutation.

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import math

# Fonction pour calculer le nombre de chemins possibles en fonction du nombre de villes
def nombre_chemins_possibles(n):
    return math.factorial(n - 1) // 2

# Créer des listes pour stocker les nombres de villes et les nombres de chemins possibles
nombres_villes = []
nombres_chemins = []

# Générer les données pour différentes valeurs de n
for n in range(2, 11):
    nombres_villes.append(n)
    nombres_chemins.append(nombre_chemins_possibles(n))

# Convertir les listes en tableaux numpy
x_data = np.array(nombres_villes)
y_data = np.array(nombres_chemins)

# Définir la fonction exponentielle pour la régression
def fonction_exponentielle(x, a, b):
    return a * np.exp(b * x)

# Effectuer la régression exponentielle
params, covariance = curve_fit(fonction_exponentielle, x_data, y_data)

# Obtenir les paramètres ajustés
a, b = params

# Créer une plage de valeurs pour le tracé de la courbe ajustée
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = fonction_exponentielle(x_fit, a, b)

# Tracer le graphique avec la courbe ajustée
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, marker='o', linestyle='-', label='Données')
plt.plot(x_fit, y_fit, 'r-', label='Régression exponentielle')
plt.xlabel("Nombre de villes")
plt.ylabel("Nombre de chemins possibles")
plt.grid(True)

# Ajouter le titre en bas avec "Figure 3 :"
title = "Figure 3 : Évolution du nombre de permutations en fonction du nombre de villes"
plt.figtext(0.5, 0.01, title, ha="center", fontsize=10, va="bottom")

# Afficher le graphique
plt.legend()
plt.show()
````

##**Résolution du problème par une méthode combinée: Recherche par tabou + Recuit Simulé**

###**Pourquoi choisir une telle méthode.**

Dans les méthodes de recherche dites heuristiques, de nombreux algorithmes on vut le jour: [1] [2]
 - la Brute Force qui consiste à calculer toutes les possibilités et récupérer la meilleure de toute devenant forcément la solution optimale.
 - la méthode Greedy qui consiste à dessiner les cercles les plus petits possibles entre les noeuds de solution.
 - Opt qui consiste à effectuer des échanges à partir d'une solution initiale.
 - Le Recuit Simulé qui utilise la condition de metropolis pour trouver la solution optimale
 - La Recherche par Tabou qui bani des solutions déjà visitées durant une certaine période pour explorer l'étendu des possibles.
 - La Colonie de Fourmi qui simule le déplacement d'une colonie de fourmi vers la meilleure solution.
 - etc.

Le Recuit Simulé a beaucoup marqué la résolution du TSP par son efficacité à sortir des minima locaux tout comme la Recherche par Tabou. Néanmoins, tout deux possèdent leurs faiblesses: [3]

- Pour le Recuit Simulé, un nombre conséquent d'itération est nécéssaire en plus d'un ajustement des hyper-paramètres très fin en plus de recherche répétée car aucune solution n'est enregistrée. Seul le "hasard" joue.
- Pour la Recherche par tabou, manque de contrôle et la condition d'interdiction de mouvement doit être particulièrement bien installé. De plus, la première solution impacte énormément le résultat final qui ne peut jamais converger.

Intuitivement, on peut se dire qu'en effectuant une combinaison de ces deux algorithmes on peut avec les forces de l'un compenser les faiblesses de l'autre et vice-versa.

C'est donc l'objectif de ce travail de recherche et de création. Développer un algorithme hybride de Recherche par Tabou et de Recuit Simulé.

###**Formalisme mathématiques**

Le Travelling Salesman Problem peut se résumer à la recherche de la minimisation de la fonction suivante:
**∑ ∑ xᵢ ⱼ lᵢ ⱼ** avec **(i,j) ∈ V** avec **V** l'ensemble des indices des viles.

Ici, **lᵢ ⱼ** correspond à la distance entre les villes i et j et xᵢⱼ correspond à la décision d'aller ou non entre i et j. **xᵢ ⱼ** est soit égale à 0 ou 1 et **lᵢ ⱼ** au nombre de connection possible entre deux villes du jeux de donnée.

On considerera une approche symmétrique où **lᵢ ⱼ** = **lⱼ ᵢ**



####**Formalisme de la partie Recherche par Tabou**


Longueur d'epoch ou elen est obtenu aussi de manière empyrique, elle permet de donner une limite aux itérations et de calculer le nombre de mouvement interdit et le nombre de solution à générer pour l'algorithme 2-opt-swap:

$$
\boldsymbol{{elen} = \frac{5600 \times (N^{0.4}) \times (1.27 + \gamma^{4.11}) \times \left( 4.72 \times 10^{-11} \times (\gamma + 0.1) + N^{-2.81} \right)}{(22.1 + \gamma^{4.11}) \times \left( 1.42 \times 10^{-11} + N^{-2.81} \right)}}
$$

Le gamma permet de mesurer le degré d'irrégularité et dispersion d'une série numérique. Dans notre cas, il a été observé que ce facteur joue un role important dans l'influence du nombre de tabou, la probabilité d'accepter une solution et le nombre de solutions intial généré par l'algorithme d'échange 2-opt: [3]

$$
\boldsymbol{\gamma} = \frac{\boldsymbol{nb\_town \times np.std(lᵢⱼ)}}{\boldsymbol{∑lᵢⱼ}}
$$


Le nombre de mouvement interdit ou tabu_tenure est décrit par la relation empyrique suit:

$$
\boldsymbol{\text{tabu_tenure} = \frac{elen^{0.6}}{3.5}}
$$



####**Formalisme pour l'algorithme d'échange 2-opt**

On utilisera cette algorithme pour initier une série de première solutions pour savoir le nombre de solutions dans cette série, on utilise la formule suivante:

$$
\boldsymbol{C\_N} = \frac{\boldsymbol{2800 \times (N^{1.1})}}{\boldsymbol{elen}}
$$


####**Formalisme pour le Recuit Simulé**

La probabilité d'accepter une solution possède une formule aussi:

$$
\boldsymbol{probability} = \boldsymbol{-2.46 \times N \times \frac{(new\_distance - current\_distance)}{temperature \times new\_distance \times (3.7 + (\gamma^{1.1}))}}
$$

avec new_distance: la distance générée et current_distance: la distance actuelle

On cherche à résoudre cette équation:

$$
\boldsymbol{probability < exp\left(\frac{-\delta}{temperature}\right)}
$$

````python
#functions
def random_generation_path(matrix):
    # generates a route randomly
    random_town = np.random.choice(matrix.index, replace=False, size=len(matrix))
    return random_town
````

La première solution sera générée de manière aléatoire avec cette fonction.
On tire au hasard un numéro de l'index des villes sans shanghai, on repète cette action nb_town-1 fois sans remise. Un première solution est créée.

````python
def generate_path_distance(initialise, matrix, random_town):
    # initialization
    arc_l = []
    dist = great_circle(initialise['Coordinates'], matrix.iloc[random_town[0]]['Coordinates']).km
    arc_l.append(dist)
    # total distance calculation and between towns calculation
    for i in range(1, len(random_town)):
        arc = great_circle(matrix.iloc[random_town[i - 1]]['Coordinates'], matrix.iloc[random_town[i]]['Coordinates']).km
        arc_l.append(arc)
        dist = dist + arc
    return random_town, int(dist), arc_l
````

La fonction generate_path_distance permet de calculer:
 - La distance totale, en ayant pensé à inclure shanghai, à partir des coordonées géographiques des villes
 - La distance entre deux villes
 - Rend le chemin utilisé, la distance associée et une liste receuillant les distances entre les villes traversées

````python
def sort_solution(candidates, initialise, matrix):
    # initialization
    min_arc = 0
    min_dist = np.inf
    min_route = []
    # sorting the min dist from the 2-opt-swap algo
    for route in candidates:
        route, dist, arc_l = generate_path_distance(initialise, matrix, route)
        # testing conditions
        if dist < min_dist:
            #saving results
            min_dist, min_route, min_arc = copy.deepcopy(dist), copy.deepcopy(route), copy.deepcopy(arc_l)
    return min_route, min_dist, min_arc
````

Cette fonction permet de donner le minimum des solutions engendrées par l'algorithmes 2-opt-swap utilisés plus tard.

````python
def get_gamma(l_arc, dist):
    # calculating degree of irregularity
    gamma = (nb_town * np.std(l_arc))/ (dist)
    return gamma
````

Permet de calculer la constante gamma à partir des distances entres les villes utilisées et le nombre de villes

````python
    def swap_2_opt(route):
        # applying 2-opt algorithm
        list_candidates = []
        # generating solutions for a temperature
        for i in range(C_N):
            # creating nodes for segmentation
            node1 = 0
            node2 = 0
            # giving values to the nodes
            while node1 == node2:
                node1 = randint(1, len(route)-1)
                node2 = randint(1, len(route)-1)
                # swap procedure
                if node1 > node2:
                    swap = node1
                    node1 = node2
                    node2 = swap
                # reorganizing the route
                tmp = route[node1:node2]
                tmp_bis = route[:node1] + tmp[::-1] + route[node2:]
                # saving results
                list_candidates.append(tmp_bis)
        return list_candidates
````

L'algorithme d'échange 2-opt permet de réorganiser une suite.

Il est basé sur le hasard. On prend au hasard 2 noeuds de l'ordre de villes initial puis on s'assure que ces valeurs soient bien en ordre croissant de leur indice.

Ensuite, on créer une nouvelle liste avec ces noeuds en combinant les trois morceaux: le chemin allant jusqu'au première noeud, l'inverse des valeurs entre les 2 noeuds de l'ordre initial puis on ajoute le chemin allant du deuxième noeuf jusqu'à la fin.

````python
def main(initialise=shanghai, matrix=town_without_shanghai):
    global alpha, temp_i  # hyper-parameters
    global deny, accept, tabu_list, tabu_index, stop  # algorithms variables
    global best_path, best_dist, old_best_path  # best result variables

    # initialisation by generating a random path
    random_town = random_generation_path(matrix)
    route, dist, arc_l = generate_path_distance(initialise, matrix, random_town)
    route = route.tolist()

    gamma = get_gamma(arc_l, dist)

    def get_hp():
        elen = int((5600*(nb_town**0.4)*(1.27+(gamma**4.11))*(4.72*(10**-11)*(gamma+0.1)+(nb_town**-2.81)))/((22.1+(gamma**4.11))*(1.42*(10**-11)+(nb_town**-2.81))))  # epoch length
        tabu_tenure = int(((elen**0.6)/3.5))  # max number of tabu moves
        C_N = int(((2800*(nb_town**1.1))/elen))  # number of solutions proposed per temperature
        return elen, tabu_tenure, C_N

    elen, tabu_tenure, C_N = get_hp()  # getting the rest of hyper-parameters

    """
    creating the functions we need to apply the combinated method of research
    """

    def swap_2_opt(route):
        # applying 2-opt algorithm
        list_candidates = []
        # generating solutions for a temperature
        for i in range(C_N):
            # creating nodes for segmentation
            node1 = 0
            node2 = 0
            # giving values to the nodes
            while node1 == node2:
                node1 = randint(1, len(route)-1)
                node2 = randint(1, len(route)-1)
                # swap procedure
                if node1 > node2:
                    swap = node1
                    node1 = node2
                    node2 = swap
                # reorganizing the route
                tmp = route[node1:node2]
                tmp_bis = route[:node1] + tmp[::-1] + route[node2:]
                # saving results
                list_candidates.append(tmp_bis)
        return list_candidates

    def tm_verify_length():
        # verify if the length of tabu_list is respected
        if len(tabu_list) > tabu_tenure:
            return False

    def tm_test(dist, route):
        global tabu_list, tabu_index
        # testing tabu moves
        for i in range(len(tabu_list)):
            if route[tabu_index[i]] == tabu_list[i]:
                return False
        return True

    def tm_update(old_best, actual_best):
        global tabu_list, tabu_index
        # updating the tabu moves list
        for i in range(len(actual_best)):
            if old_best[i] != actual_best[i]:
                tabu_list.append(old_best[i])
                tabu_index.append(i)
        # reset tabu list if stopping criteria is reached
        if len(tabu_list) > tabu_tenure:
            while len(tabu_list) - tabu_tenure != 0:
                if tabu_list != []:
                    tabu_list.pop(0)
                    tabu_index.pop(0)
        return tabu_list, tabu_index

    def get_probability(new, current):
        global temp_i
        # getting probability for the s_a algorithm logic
        if new - current <= 0:
            probability = -2.46 * nb_town * (new - current) / (temp_i * new * (3.7 + (gamma ** 1.1)))
        else:
            probability = 1
        return probability

    def sa(new_dist, current_dist, alpha):
        global temp_i
        # s_a algorithm
        delta_distance = new_dist - current_dist
        probability = get_probability(new_dist, current_dist)
        if probability < np.exp((-delta_distance) / temp_i):
            return True
        else:
            return False

    # creation of the current solution
    current_dist = copy.deepcopy(dist)
    current_path = copy.deepcopy(route)

    # loop
    while temp_i > temp_f:
        print('--------')
        print('temperature actuelle:',temp_i)
        num_iter = int(elen / 10)
        for i in range(num_iter):
            print('--------')
            print('iteration numero:', i, 'sur', num_iter)
            # applying 2-opt-swap algo and selecting the best result and calculating distance
            candidates = swap_2_opt(route)
            route, dist, arc_l = sort_solution(candidates, initialise, matrix)

            # recalculating hyper-parameters
            gamma = get_gamma(arc_l, dist)
            elen, tabu_tenure, C_N = get_hp()  # hyper-parameters

            # aspiration criteria
            if dist < best_dist:
                # updating results
                old_best_path = copy.deepcopy(best_path)
                best_dist = copy.deepcopy(dist)
                best_path = copy.deepcopy(route)
                # updating tabu_list as results are being improved
                tabu_list, tabu_index = tm_update(old_best_path, best_path)

            # t_s algorithm
            elif not tm_verify_length():
                if tm_test(dist, route):
                    # s_a algorithm
                    if sa(dist, current_dist, alpha):
                        # updating results
                        current_dist, current_path = dist, route
            # checking if results are better, aspiration criterion
            if current_dist < best_dist:
                # updating results
                old_best_path = copy.deepcopy(best_path)
                best_dist = copy.deepcopy(current_dist)
                best_path = copy.deepcopy(current_path)
                # updating tabu_list as results are being improved
                tabu_list, tabu_index = tm_update(old_best_path, best_path)

            # printing result while looping
            print('--------')
            print('meilleure distance actuelle:', best_dist)
            print('nombre de tabou:', len(tabu_list))

        temp_i *= alpha

    return best_dist, best_path
````

## **Résolution du TSP par la méthode de la colonie de fourmis**

###**Pourquoi avoir un algorithme basé sur le comportement des fourmis**

Avant de détailler le formalisme mathématique de la colonie de fourmis, nous allons vous présenter les caractéristiques fascinantes des fourmis.

Les fourmis sont des animaux sociaux ayant un mode de communication très développé. En effet, elles utilisent des substances chimiques nommées phéromones pour communiquer entre elles. Ces substances sont des mélanges d'hydrocarbures que la fourmi sécrète et dépose sur son chemin pour attirer les autres fourmis.

Il est important de noter que les phéromones ont la particularité de s'évaporer plus ou moins vite avec le temps. De ce fait, si le chemin emprunté par la première fourmi n'est pas revisité par d'autres fourmis, alors les phéromones s'évaporeront et les fourmis n'auront plus aucune attirance pour ce chemin. Il sera donc voué à disparaître.

Un autre phénomène intéressant chez les fourmis est leur façon d'optimiser leur déplacement et de recruter en masse leurs congénères afin d'effectuer une tâche. L'exemple le plus parlant est celui de la recherche de nourriture pour assurer la survie de la colonie.

Dans ce cas de figure, une première fourmi part de son nid (point N sur la figure) pour aller chercher de la nourriture (point F sur la figure). Dès qu'elle aura trouvé une source de nourriture, elle va revenir jusqu'à son nid en prenant un chemin quelconque tout en déposant des phéromones sur ce dernier. Les autres fourmis seront attirées par les phéromones et vont alors aller jusqu'à cette source de nourriture via différents chemins. Cependant, au fur et à mesure du temps, les fourmis vont privilégier le chemin le plus court. Elles vont donc déposer énormément de phéromones sur ce dernier pour que les autres fourmis prennent uniquement ce chemin. Les autres chemins finiront par disparaître en raison de l'évaporation des phéromones. Les fourmis auront donc optimisé leur trajet.

###**Formalisme mathématique**

Il existe plusieurs type d’algorithme inspiré du comportement des fourmis. Nous ne traitons que de l’algorithme Ant Colony System qui est une amélioration de l'algorithme Ant System.
Ant System fût l’un des premiers à résoudre le problème du voyageur de commerce.


####**Construction d'une solution**

Lorsque les fourmis doivent se déplacer de leur ville initiale à une autre ville, elle utilise des probabilités dépendantes de deux facteurs pour savoir où aller.
Il s’agit de la quantité de phéromones déposée sur les villes et la distance entre ces villes.
Les probabilités sont calculées à partir de la loi suivante :


$$P_{ij}(t) = \frac{[\tau_{ik}(t)]^\alpha \cdot [\eta_{ik}]^\beta}{\sum_{k \in \text{disponibles}} [\tau_{ij}(t)]^\alpha \cdot [\eta_{ij}]^\beta}$$


$$P_{ij}(t)$$ représente la probabilité pour que cette ville soit la prochaine ville où la fourmi doit aller

$$[\tau_{ik}(t)]$$ représente la quantité de phéromone sur le chemin entre deux villes

$$[\eta_{ik}]$$ représente la distance entre deux villes

$$k$$ représente une fourmi du système

$$disponible$$ correspond à l'ensemble des villes non visitées

$$alpha$$ correspond à l'influence des phéromones

$$béta$$ à l'influence de la distance


Le numérateur de cette fraction permet de prendre en compte les facteurs importants de la probabilité (phéromones et distance).
Le dénominateur sert à normaliser les probabilités.

####**Evaporation des phéromones**

A chaque déplacement des fourmis, les phéromones déposés par leurs prédécesseurs s'évaporent selon la formule :

$$\tau_{ij}(t) = (1 - \rho) \cdot \tau_{ij}(t-1)$$


ici rho correspond à la vitesse d'évaporation des phéromones.

####**Mise à jour des phéromones**

Une fois que la fourmi aura emprunté un chemin, elle va déposer une quantité de phéromones. Cette quantité est répresentée par la formule suivante et tient compte des phéromones qui se sont évaporées :

$$\tau_{ij}(t) = (1 - \rho) \cdot \tau_{ij}(t-1) + \sum_{k \in \text{ville}} \Delta\tau_{ik}$$


$$\Delta\tau_{ij} = \frac{1}{L_{\text{fourmi}}}$$

Ici L correspond à la longueur du chemin (distance entre les villes)

####**Selection du chemin optimal**

Avec le temps, les fourmis vont parcourir l'ensemble des villes tout en déposant des phéromones. Cela influencera le choix des autres fourmis, et le chemin le plus optimal apparaîtra, donnant ainsi la meilleure solution au problème.

Plus il y a de phéromones, plus les fourmis emprunteront ce chemin. Les chemins qui ne sont plus empruntés verront leur quantité de phéromones disparaître.

````python
from geopy.distance import great_circle
def distance_haversine(a,b):
  return great_circle(a, b).km
````

Depuis le dataframe coordinates, nous allons regrouper les coordonnées des villes dans la variable cities. Chaques villes possèdent les colonnes 0 et 1 qui correspondent respectivement à leur latitude et leur longitude. La variable totales_villes correspond au total du nombre de ville.

On initie ensuite une matrice vide qui servira à stocker les distances entre deux villes qu'on nommera "distances". Grâce à la fonction distance_haversine, nous allons calculer la distance entre toutes les villes l'une par rapport à l'autre. L'ensemble des distances sont ensuite stockées dans la matrice distance.
````python
distances = np.zeros((totales_villes, totales_villes))
for i in range(totales_villes):
    for j in range(i + 1, totales_villes):
        distances[i][j] = distance_haversine(villes[i], villes[j])
        distances[j][i] = distances[i][j]
````

On va maintenant initier les hyperparamètres de la méthode ACO à savoir :

- le nombre de fourmi (chaque fourmi devra trouver une solution pour avoir la distance la plus courte en commençant par Shanghai)

- la vitesse d'évaporation des phéromones (il s'agit de la vitesse d'évaporation sur chaques connexions entre deux villes)

- le taux de dépot des phéromones (il s'agit de la quantité de phéromone déposé par les fourmis pour chaque connexion entre deux villes)



````python
nb_fourmis = 100
nb_iterations = 150
alpha = 1.0
beta = 3.0
vitesse_evaporation = 0.5
taux_de_dépot_pheromones = 1.0
````

On commence par initialiser une matrice correspondant aux phéromones sur chacune des connexions entre deux villes. Cette matrice, appelé pheromones, stockera les phéromones déposées par les fourmis.


````python
pheromones = np.ones((totales_villes, totales_villes))
````

Pour chaque itération parmi le nombre d'itération défini en hyperparamètre, on crée une liste vide appelé "routes_fourmis" qui servira à stocker le chemin parcouru par chaque fourmi.

Les fourmis commencent à Shanghai puis elles vont créer deux listes :
- une pour les villes déjà visitées : "villes_visitées"
- une autre pour les villes qui restent à visiter : "villes_restantes"

Pour savoir quelle ville les fourmis doivent visiter, elles utilisent des probabilités (voir le formalisme mathématique pour plus d'explication) qui dépendent des phéromones déposés entre chaque chemin de ville ainsi que de la distance entre les villes.

Les probabilités sont stockés dans la liste initialement vide "probabilites" et sont ensuite transformées pour avoir des valeurs entre 0 et 1.

En fonction de ces probabilités, la prochaine ville est sélectionnée et les listes "villes_visitées" et "villes_restantes" sont mis à jour.

Dès que les fourmis ont terminé leur trajet, les phéromones sont mis à jour en fonction des hyperparamètres vitesse_evaporation et taux_de_dépot_pheromones. Plus le chemin entre deux villes est court, plus les phéromones sont élevés. Ce sont les phéromones qui determineront quel chemin les fourmis devront prendre.

````python
for iteration in range(nb_iterations):
    routes_fourmis = []

    for fourmi in range(nb_fourmis):
        ville_actuelle = 0
        villes_visitées = [ville_actuelle]
        villes_restantes = list(range(1, totales_villes))

        while villes_restantes:
            probabilites = []
            for ville in villes_restantes:
                prob = (pheromones[ville_actuelle][ville] ** alpha) * ((1.0 / distances[ville_actuelle][ville]) ** beta)
                probabilites.append(prob)

            probabilites = np.array(probabilites)
            probabilites /= probabilites.sum()
            prochaine_ville = np.random.choice(villes_restantes, 1, p=probabilites)[0]

            villes_visitées.append(prochaine_ville)
            villes_restantes.remove(prochaine_ville)
            ville_actuelle = prochaine_ville

        routes_fourmis.append(villes_visitées)

    pheromones *= (1.0 - vitesse_evaporation)
    for route in routes_fourmis:
        distance_trajet = sum(distances[route[i]][route[i + 1]] for i in range(totales_villes - 1))
        distance_trajet += distances[route[-1]][route[0]]
        for i in range(totales_villes - 1):
            pheromones[route[i]][route[i + 1]] += taux_de_dépot_pheromones / distance_trajet
        pheromones[route[-1]][route[0]] += taux_de_dépot_pheromones / distance_trajet
````

Nous allons maintenant créer deux variables :

- Une autre qui correspondra à la liste des villes visitées pour avoir la plus petite distance parcourue.

##**Approche bayésienne**

L'approche bayésienne est une méthode d'optimisation qui permet de trouver l'extremum global d'une fonction continue, potentiellement non dérivable et réelle avec un nombre de dimensions relativement faible ($d<20$).
A première vue, cette approche s'applique naturellement mal au problème TSP, car celui-ci est discretisé sur la ponctualité de chaque villes (et de sa représentation sous forme de coordonnées spatiales). Nous allons donc tenter de contourner le problème en utilisant des approches suggérées par la littérature.
Il n'existe pas de littérature disponible sur Internet qui traite du TSP avec un algorithme bayésien, alors la solution présentée sera inspirée de méthodologies adjacentes dont il est possible de tirer profit.

###**Formalisme**

- On cherche le $\max_{x \in A} f(x)$
- Observation initiale : $ x_1, x_2, \ldots, x_{n_0} \in \mathbb{R}, [f(x_1), \ldots, f(x_{n_0})]$
- Prior initial : $f(x_{1:n_0}) \sim \mathcal{N}(\mu_0(x_{1:n_0}), \Sigma_0(x_{1:n_0}, x_{1:n_0}))$
- Fonction de noyau $RBF(x, x_0) = \exp(-\gamma \|x - x_0\|^2)$
- Gain d'information attendu : $EI(x; \xi) = (\mu - f(x^\star) - \xi) \Phi\left(\frac{\mu - f(x^\star) - \xi}{\sigma}\right) + \sigma\phi\left(\frac{\mu - f(x^\star) - \xi}{\sigma}\right)
$

Ici, $x^\star$ est la meilleure solution à une étape $n$, $\xi$ décrit le biais "exploration / exploitation", $\phi$ représente la fonction noyau, avec $mu$ et $\sigma$ ses hyperparamètres dans le cas d'une distribution gaussienne, et $\Phi$ la quantité $\Phi(x) = \int_{-\infty}^x RBF(t) \, dt$

###**Comment fonctionne-t-elle ?**

On crée un procédé gaussien sur l'ensemble de la fonction $f$. On crée une observation primaire avec un nombre prédéfini d'échantillons $n_0$. A chaque fois qu'une observation est faite, elle apporte un gain d'informations qui va soit isoler un extremum, soit améliorer le processus.

###**Principe sous forme d'algorithme :**

Placez une priorité (prior) de processus Gaussien sur $f$ selon la fonction de noyau (kernel).
Observez $f$ en $n_0$ points selon une conception expérimentale initiale d'occupation de l'espace (e.g. la fonction d'acquisition). Définissez $n = n_0$.
#####Tant que $n \leq N$ :
   - Mettez à jour la distribution de probabilité postérieure sur $f$ en utilisant toutes les données disponibles.
   - Laissez $x_n$ être un maximiseur de la fonction d'acquisition sur $x$, où la fonction d'acquisition est calculée en utilisant la distribution postérieure courante.
   - Observez $y_n = f(x_n)$.
   - Incrémentez $n$.

Fin du tant que.

Renvoyer une solution : soit le point évalué avec la plus grande $f(x)$, soit le point avec la plus grande moyenne *a posteriori*.

Cette méthode repose sur une fonction, un noyau, qui va approximer la densité de probabilité entre deux observations. On choisit souvent la RBF disponible sur $sklearn$, mais des alternatives existent et présentent différents avantages selon la situation. Ces fonctions auront toutefois besoin d'un choix d'hyperparamètres, qui est laissé (par définition) à l'utilisateur.

La manière dont sera réalisé l'échantillonnage est aussi régi par une fonction, d'acquisition. Cette fonction doit prévoir le meilleur échantillonnage en se basant sur l'attente du meilleur gain d'information possible. C'est ici qu'entre en jeu le dilemme exploration (gagner le plus d'information possible sur la fonction) vs. exploitation (trouver un point d'intérêt ayant la plus fort probabilité d'être un extremum). Ainsi, les fonctions utilisées, qui peuvent être basées sur une montée de gradient d'information ou un mécanisme d'entropie, dépendent aussi d'hyperparamètres à la merci de l'utilisateur.

Voici ensuite la fonction qui permet à son tour de calculer la distance entre deux points, ou même entre une séquence de points à travers la création d'un dictionnaire

````python
import csv
from geopy import distance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import math
import copy



def calculate_distance(path):
    total_distance = 0
    with open('/content/drive/MyDrive/IA/pylearn-main/TeamWork/geonames-all-cities-with-a-population-1000.csv', newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile, delimiter=';')
        # Skip the header row
        next(reader)
        # create a dictionary to store the coordinates of each geoname
        coords_dict = {}
        for row in reader:
            geoname_id = row[0]
            lat, lon = row[-1].split(',')
            coords_dict[int(geoname_id)] = (float(lat), float(lon))

    for i in range(len(path)-1):
        # get the coordinates of the current and next geonames
        current_geoname = path[i]
        next_geoname = path[i+1]
        current_coord = coords_dict[current_geoname]
        next_coord = coords_dict[next_geoname]
        # calculate the distance between the current and next geonames
        dist = distance.distance(current_coord, next_coord).km
        total_distance += dist

    return total_distance
````

Voici le module utilisé pour optimiser la fonction calculate_distance().

````python
!pip install bayesian-optimization
from bayes_opt import BayesianOptimization

def optimize():
    # Define the bounds for the city order
    bounds = {'city{}'.format(i): (1, 24338) for i in range(1, 11)}

    # Define the objective function for Bayesian optimization
    def objective_function(**kwargs):
        city_order = [int(kwargs['city{}'.format(i)]) for i in range(1, 11)]
        distance, _ = calculate_distance(city_order)
        return -distance

    # Generate a random initial city order
    initial_city_order = list(range(1, 11))
    random.shuffle(initial_city_order)
    initial_guess = {'city{}'.format(i): initial_city_order[i-1] for i in range(1, 11)}

    # Perform Bayesian optimization to minimize the distance
    optimizer = BayesianOptimization(f=objective_function, pbounds=bounds, random_state=1)
    optimizer.maximize(init_points=initial_guess, n_iter=50)

    # Print the optimized city order and distance
    optimized_city_order = [int(optimizer.max['params']['city{}'.format(i)]) for i in range(1, 11)]
    optimized_distance = -optimizer.max['target']
    print('Optimized city order:', optimized_city_order)
    print('Optimized distance:', optimized_distance)
````

````python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# Sample list of indexes (cities)
indexes = list(int(reader[0]))

# Number of cities in each path for starters
path_length = 25

# Number of random paths you want to generate
num_paths = 10E3  # This will decide the amount of resolution we want to give our GP function

random_paths = []

for _ in range(num_paths):
    # Shuffle the indexes randomly to create a random path
    random.shuffle(indexes)

    # Take the first 'path_length' indexes to create a path
    path = indexes[:path_length]

    random_paths.append(path)

# random_paths now contains a list of random paths of 10 cities each

paths = random_paths  # List of paths
distances = [calculate_distance(path) for path in paths]  # Corresponding calculated distances

# Define the kernel
kernel = 1.0 * RBF(length_scale=1.0)

# Create and fit the Gaussian Process model
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp.fit(paths, distances)
````
